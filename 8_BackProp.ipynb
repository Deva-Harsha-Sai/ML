```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df=pd.read_csv("/content/drive/MyDrive/Naive bayes/Document.csv",names=["text","label"])
```

```python
df["numlabel"]=df.label.map({"pos":1,"neg":0})
```

```python
df
```

```python
#Text Preprocessing
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = []
for i in range(len(df)):
    #Lowercase
    data = df['text'][i].lower()

    #Create a list of strings using string.split() method
    data = data.split()

    # Prefer Lemmatization over Stemming
    #data = [ps.stem(word) for word in data if not word in stopwords.words('english')]
    data = [wl.lemmatize(word, pos='v') for word in data if not word in stopwords.words('english')]
    data = ' '.join(data)
    corpus.append(data)
    print(i, end = ' ')
```

```python
corpus
```

```python
df['processed_text'] = corpus
df
```

```python
cv = CountVectorizer()
X_cv = cv.fit_transform(corpus)
X_cv = X_cv.toarray()
X_cv
```

Train - Test - Split

```python
X_train_cv, X_test_cv, y_train_cv, y_test_cv = train_test_split(X_cv, list(df['numlabel']), test_size = 0.2, random_state = 0)
```

MLP (Multi Layer Perceptron) Classifier

```python
model_mlp_cv = MLPClassifier(activation = 'logistic')
model_mlp_cv.fit(X_train_cv, y_train_cv)
y_pred_mlp_cv = model_mlp_cv.predict(X_test_cv)
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy using Count Vectorizer and Multilayer Perceptron: ", accuracy_score(y_test_cv, y_pred_mlp_cv))
```

```python
print('Confusion Matrix: \n', confusion_matrix(y_test_cv, y_pred_mlp_cv))
```
